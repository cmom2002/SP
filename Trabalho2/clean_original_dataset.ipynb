```python
import random
import pandas as pd
```

```python
path = "dataset/netflix_titles.csv"
file = pd.read_csv(path)
df = pd.DataFrame(file)
display(df)
```

```python
for i in range(len(df)):
    tipo = df["type"][i]
    if tipo == "TV Show":
        index_line = df.index[df["type"] == tipo].tolist()[0]
        df = df.drop(index_line)

df = df.drop("type", axis=1)       
df.to_csv("dataset/netflix_1.csv", index=False)
```

```python
path = "dataset/netflix_1.csv"
file = pd.read_csv(path)
df = pd.DataFrame(file)
display(df)
```

```python

for i in range(len(df)):   
    rating = random.randint(1, 5)
    df["rating"][i] = rating
    genero = df["listed_in"][i]
    if "," in genero:
        df["listed_in"][i] = genero.split(",")[0]

df.to_csv("dataset/netflix_final.csv", index=False)
display(df)
```
