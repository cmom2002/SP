```python
import pandas as pd
import random
```

```python
path = "dataset/netflix_final.csv"
file = pd.read_csv(path)
df = pd.DataFrame(file)

display(df)
```

```python
n_lines = 100000

n_filmes = n_lines * 1.5
n_realizador = 20

nome_filme = ["filme" + str(i) for i in range(1, int(n_filmes))]
nome_realizador = ["realizador" + str(i) for i in range(1, n_realizador)]
tipo_genero = df["listed_in"].unique()
```

```python
def commun_data(df_aux1, df_aux2):
    array_commun = df_aux1["Filme"].isin(df_aux2["Filme"])
    
    count = array_commun.sum()
    print(count)
```

```python

def new_dataset():
    df_aux1 = None
    df_aux2 = None
    for i in range(2):
        array_filme = []
        array_realizador = []
        array_duracao = []
        array_genero = []
        array_lancamento = []
        array_rating = []
        for j in range(n_lines):
            while(1):
                f = random.choice(nome_filme)
                if f not in array_filme:
                    array_filme.append(f)
                    break
            array_realizador.append(random.choice(nome_realizador))
            array_duracao.append(str(random.randint(30, 240)) + " min")
            array_genero.append(random.choice(tipo_genero))
            array_lancamento.append(random.randint(1960, 2023))
            array_rating.append(random.randint(1, 5))

        d = {"Filme": array_filme, "Realizador": array_realizador, "Duração": array_duracao, "Género": array_genero, "Ano lançamento": array_lancamento, "Rating": array_rating}
        df_d = pd.DataFrame(d)
        
        if i == 0:
            df_aux1 = df_d
            df_d.to_csv("dataset/PartA/dataset" + str(n_lines)+ ".csv", index=False)
        elif i == 1:
            df_aux2 = df_d
            df_d.to_csv("dataset/PartB/dataset" + str(n_lines)+ ".csv", index=False)
    
    commun_data(df_aux1, df_aux2)


new_dataset()
```

```python
def subdivide_dataset(path1):
    path = path1 + "dataset100000.csv"
    file = pd.read_csv(path)
    df = pd.DataFrame(file)
    sd = 20000
    for i in range(5):
        df2 = pd.DataFrame(df.head(sd))
        df2.to_csv(path1 + "dataset" + str(sd) + ".csv", index=False)
        sd += 20000

subdivide_dataset("dataset/PartA/")
subdivide_dataset("dataset/PartB/")
```

```python
patha = "dataset/PartA/dataset20000.csv"
pathb = "dataset/PartB/dataset20000.csv"
filea = pd.read_csv(patha)
dfa = pd.DataFrame(filea)
fileb = pd.read_csv(pathb)
dfb = pd.DataFrame(fileb)
commun_data(dfa, dfb)
```
